指标	2022年	2021年	2020年	2019年	2018年
国民总收入(亿元)	1197250.4	1141230.8		1005451.3	983751.2	915243.5
国内生产总值(亿元)	1210207.2	1149237.0	1013567.0	986515.2	919281.1

```python
import pandas as pd

data = [
    ['国民总收入(亿元)', 1197250.4, 1141230.8, 1005451.3, 983751.2, 915243.5],
    ['国内生产总值(亿元)', 1210207.2, 1149237.0, 1013567.0, 986515.2, 919281.1],
]

columns = ['指标', '2022年', '2021年', '2020年', '2019年', '2018年']

df = pd.DataFrame(data, columns=columns)

df.to_csv('指标数据.csv', index=False, encoding='utf-8-sig')

print('已写入 指标数据.csv')
print(df)
```

```python
import matplotlib.pyplot as plt

# 读取已保存的 CSV（含 BOM）
df_csv = pd.read_csv('指标数据.csv', encoding='utf-8-sig')

# 将“指标”设为列索引并转置，方便按年份比较各指标
df_plot = df_csv.set_index('指标').T
df_plot.index.name = '年份'

# 绘制直方图（这里使用分组柱状图显示每年各指标数值）
ax = df_plot.plot(kind='bar', figsize=(10,6))
ax.set_ylabel('数值（亿元）')
ax.set_title('各年指标比较')
plt.xticks(rotation=0)
plt.tight_layout()
plt.rcParams['font.sans-serif']=['Microsoft YaHei']
plt.show()
```
