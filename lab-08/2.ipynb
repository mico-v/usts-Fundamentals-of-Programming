```python
import pandas as pd
import matplotlib.pyplot as plt
# 读取之前保存的 CSV（与 notebook 同目录）
df = pd.read_csv('地区2022_gdp.csv', encoding='utf-8')
df.columns = df.columns.str.strip()
# 使用全部数据绘制饼图（不合并为其他）
labels = df['地区'].tolist()
sizes = df['2022年地区生产总值（亿元）'].tolist()
# 中文字体和负号设置（按需会回退至系统默认字体）
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False
fig, ax = plt.subplots(figsize=(10,10))
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
ax.set_title('2022年各地区地区生产总值（占比）')
fig.savefig('2022_gdp_pie.png', dpi=200, bbox_inches='tight')
print('已保存饼图: 2022_gdp_pie.png')
plt.show()
```
