````python
#输入一组无序的整数，编程输出其中出现次数最多的整数及其出现次数。
#输入形式: 先输入整数的个数（大于等于1，小于等于100），然后在下一行输入这些整数，各整数之间以一个逗号分隔。
#输出形式: 输出出现次数最多的整数及其出现次数，两者以一个空格分隔；若出现次数最多的整数有多个，则按照整数升序分行输出。
#若输入数据不合法(第一行不是1到100之间的整数，第二行的整数格式或个数错误)，则输出"illegal input"。

#```python
from collections import Counter
try:
    n = int(input().strip())
    if n < 1 or n > 100:
        print("illegal input")
    numbers = input().strip().split(',')
    if len(numbers) != n:
        print("illegal input")
    nums = [int(num) for num in numbers]
    count = Counter(nums)
    max_freq = max(count.values())
    most_common_nums = [num for num, freq in count.items() if freq == max_freq]
    most_common_nums.sort()
    for num in most_common_nums:
        print(num, max_freq)
except:
    print("illegal input")


````

````python
#从文件中读取某城市2022年12个月份的新房销售数据，文件名为：price.csv。
# 文件格式为：每行表示一个月的销售数据，包括：月份、销售面积(单位：万米2)和均价(单位：元/米2)，数据之间用逗号分隔。例如：
#需要对销售数据进行分析，找出销量最大的月份、均价最高的月份，并计算每个月销量和房价的环比
# （环比是指本月数据与上个月的数据比较的变化百分比，由于没有上一年度12月份的数据，因此不计算1月份的环比）。
#将分析结果输出到文件名为stat.csv的文件中，文件格式为：
# 第一行是销量最高的月份和销量（如果存在并列第一，都要逐行输出），
# 第二行是均价最高的月份和均价（如果存在并列第一，都要逐行输出）。
# 然后逐行输出2月份以后的11个月的环比数据，
# 每行包括：月份、销量环比、均价环比（环比以百分比格式输出，保留小数点后1位），同一行各个数据项之间用逗号分隔。
import csv

ta = []
with open("price.csv", "r", encoding="utf-8") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        parts = line.split(",")
        if len(parts) != 3:
            continue
        ti, area, price = parts
        try:
            ta.append([ti, float(area), int(price)])
        except:
            # skip invalid lines
            continue

if not ta:
    raise SystemExit("no data")

# 找出销量（面积）最大的月份（可能并列）
max_area = max(r[1] for r in ta)
max_sale_months = [r for r in ta if r[1] == max_area]

# 找出均价最高的月份（可能并列）
max_price = max(r[2] for r in ta)
max_price_months = [r for r in ta if r[2] == max_price]

# 输出到 stat.csv
with open("stat.csv", "w", encoding="utf-8", newline='') as out:
    writer = csv.writer(out)
    # 销量最高的月份和销量
    for r in max_sale_months:
        writer.writerow([r[0], "{:.1f}".format(r[1])])
    # 均价最高的月份和均价
    for r in max_price_months:
        writer.writerow([r[0], r[2]])
    # 2月份以后的环比数据
    for i in range(1, len(ta)):
        month = ta[i][0]
        prev_area = ta[i-1][1]
        prev_price = ta[i-1][2]
        area_change = (ta[i][1] - prev_area) / prev_area * 100 if prev_area != 0 else 0.0
        price_change = (ta[i][2] - prev_price) / prev_price * 100 if prev_price != 0 else 0.0
        writer.writerow([month, "{:.1f}%".format(area_change), "{:.1f}%".format(price_change)])

````
