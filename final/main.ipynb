```python
l = {"月份":[],"销售额":[],"利润":[]}
with open("sales_data.csv","r",encoding="utf-8") as f:
    f.readline()
    for line in f:
        month,sales,profit = line.strip().split(",")
        l["月份"].append(month)
        l["销售额"].append(float(sales))
        l["利润"].append(float(profit))
print(l)
# 按照销售额排序,打印排序后的结果
sorted_data = sorted(zip(l["月份"], l["销售额"], l["利润"]), key=lambda x: x[1], reverse=True)
for month, sales, profit in sorted_data:
    print(f"月份: {month}, 销售额: {sales}, 利润: {profit}")
```
