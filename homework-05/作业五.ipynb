### 随机密码生成：
>编写一个函数，函数的功能是生成密码。函数名为CreatePassword，无输入参数。密码复杂度要求：
1. 密码长度在8-16位之间；
2. 密码必须包括至少一个大写字母、小写字母、数字、特殊字符（~!@#$%^&*?）。
>要求在主程序中调用CreatePassword函数，生成10个随机密码。

```python
import random, re, string
le = string.ascii_letters
di = string.digits
pu = "~!@#$%^&*?"
pat = re.compile(r'^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[' + re.escape(pu) + r']).{8,16}$')

def CreatePassword():
    while True:
        p = ''.join(random.choice(le + di + pu) for _ in range(random.randint(8,16)))
        if pat.match(p): return p

for _ in range(10):
    print(CreatePassword())
```

### 英文段落单词词频统计：
> 编写一个函数，函数的功能是统计英文段落中单词的词频。函数名为WordCount，有一个参数text，text包含一段英文文章。函数返回一个字典，字典的键为单词，字典的值为单词出现的次数。
> 要求在主程序中实现从键盘输入英文段落，调用WordCount函数，然后输出出现次数最多的10个单词及出现的次数。


```python
from collections import Counter
import re

def WordCount(text):
    words = re.findall(r"[A-Za-z']+", text)
    cleaned = [w.lower() for w in words]
    return dict(Counter(cleaned))

print("Enter/paste an English paragraph:")
freq = WordCount(input())
top10 = sorted(freq.items(), key=lambda kv: kv[1], reverse=True)[:10]

print("Top 10 words:")
for word, count in top10:
    print(f"{word}: {count}")
```

### 生日悖论问题：
>生日悖论是指在一个拥有23人或23人以上的群体中，至少有两个人生日相同（月和日相同）的概率会超过50%。这个概率远高于人们的直觉预期，因此被称为“悖论”，但这并非逻辑上的矛盾，而是概率结果与直觉相悖的现象。
>编写一个函数，函数的功能是通过随机函数为每个人分配一个生日，如果出现2个相同的生日，则返回True，否则返回False。函数名为hasSameBirthday，有一个参数num，num为人数。
>在主程序中通过键盘输入群体中的人数。通过10000次调用hasSameBirthday函数，统计函数返回True的次数，将出现True的次数除以10000，得到的值就是生日相同的概率。分别用人数20、23、30三种情况测试，观察概率值，发现人数大于等于23时概率值大于等于0.5。


```python
from random import randint
def hasSameBirthday(num):
    seen = set()
    for _ in range(num):
        day = randint(1, 365)
        if day in seen:
            return True
        seen.add(day)
    return False


for x in [20,23,30]:
    trueN = 0
    for _ in range(10000):
        if hasSameBirthday(x):
            trueN +=1
    print(f"当人数为{x}时概率为{trueN/10000}")
```
