### 随机密码生成：
>编写一个函数，函数的功能是生成密码。函数名为CreatePassword，无输入参数。密码复杂度要求：
1. 密码长度在8-16位之间；
2. 密码必须包括至少一个大写字母、小写字母、数字、特殊字符（~!@#$%^&*?）。
>要求在主程序中调用CreatePassword函数，生成10个随机密码。

```python
import random, re, string
le = string.ascii_letters
di = string.digits
pu = string.punctuation
pat = re.compile(r'^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[' + re.escape(pu) + r']).{8,16}$')

def CreatePassword():
    while True:
        p = ''.join(random.choice(le + di + pu) for _ in range(random.randint(8,16)))
        if pat.match(p): return p

for _ in range(10):
    print(CreatePassword())
```

### 英文段落单词词频统计：
> 编写一个函数，函数的功能是统计英文段落中单词的词频。函数名为WordCount，有一个参数text，text包含一段英文文章。函数返回一个字典，字典的键为单词，字典的值为单词出现的次数。
> 要求在主程序中实现从键盘输入英文段落，调用WordCount函数，然后输出出现次数最多的10个单词及出现的次数。


```python
from collections import Counter
import re

def WordCount(text):
    words = re.findall(r"[A-Za-z']+", text)
    cleaned = [w.strip("'").lower() for w in words if w.strip("'")]
    return dict(Counter(cleaned))

print("Enter/paste an English paragraph:")
line = input()
freq = WordCount("\n".join(line.split()))
top10 = sorted(freq.items(), key=lambda kv: kv[1], reverse=True)[:10]

print("Top 10 words:")
for word, count in top10:
    print(f"{word}: {count}")
```

```python
def hasSameBirthday(num):
    """Return True if in a group of `num` people at least two share the same birthday.
    Birthdays are uniformly random among 365 days (ignore Feb 29).
    """
    seen = set()
    for _ in range(num):
        day = random.randint(1, 365)
        if day in seen:
            return True
        seen.add(day)
    return False

def estimate_probability(num, trials=10000):
    successes = 0
    for _ in range(trials):
        if hasSameBirthday(num):
            successes += 1
    return successes / trials

# Optional: read a single number from keyboard (leave blank to skip)
try:
    s = input("Enter number of people to estimate (or press Enter to skip): ").strip()
    if s:
        n = int(s)
        print(f"Estimated probability for {n} people: {estimate_probability(n):.4f}")
except Exception as e:
    print("Input skipped or invalid:", e)

# Run the requested tests for 20, 23, 30 people
for num in (20, 23, 30):
    print(f"Num={num}: estimated probability = {estimate_probability(num):.4f}")
```
